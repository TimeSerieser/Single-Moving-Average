```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('MA.csv')
df.head()
```

```python
df.tail()
```

```python
df['Y_pred'] = 0.00
for i in range(0,38):
    df['Y_pred'][i+3] = np.mean( [ df['Y_true'][i], df['Y_true'][i+1], df['Y_true'][i+2] ])
df.head()                                
```

```python
df.tail()
```

```python
df2 = df.copy()
a = df2['Y_pred'].iloc[-1]
new_row1 = {'Month':'41', 'Y_true':'NaN', 'Y_pred':a}
df2 = df2.append(new_row1, ignore_index=True)
new_row2 = {'Month':'42', 'Y_true':'NaN', 'Y_pred':a}
df2 = df2.append(new_row2, ignore_index=True)
df2.tail(8)
```

```python
for i in range(40,41):
    df2['Y_pred'][i] = np.mean( [ df2['Y_true'][i-1], df2['Y_true'][i-2], df2['Y_true'][i-3] ])
df2.head()                                
```

```python
df['Absolute Error'] = 0.00
for i in range(3,39):
    df['Absolute Error'][i] = abs( df['Y_pred'][i] - df['Y_true'][i])
df.head()                                
```

```python
df.tail()
```

```python
df['Squared Error'] = 0.00
for i in range(3,39):
    df['Squared Error'][i] = (df['Y_pred'][i] - df['Y_true'][i])**2
df.head()                                
```

```python
df.tail()
```

```python
df['Precent Error1'] = 0.00
for i in range(3,39):
    df['Precent Error1'][i] = abs(( df['Y_true'][i] - df['Y_pred'][i] ) / df['Y_true'][i])
df.head()                                
```

```python
df.tail()
```

```python
df['Precent Error2'] = 0.00
for i in range(3,39):
    df['Precent Error2'][i] = ((df['Y_pred'][i] - df['Y_true'][i])/ df['Y_true'][i-1])**2
df.head()                                
```

```python
df.tail()
```

```python
df['Precent Error3'] = 0.00
for i in range(3,39):
    df['Precent Error3'][i] = ((df['Y_true'][i] - df['Y_true'][i-1] ) / df['Y_true'][i-1] )**2
df.head()                                
```

```python
df.tail()
```

```python
df['Error'] = 0.00
for i in range(3,39):
    df['Error'][i] = ( df['Y_pred'][i] - df['Y_true'][i] )
df.head()                                
```

```python
df.tail()
```

```python
df['Squared Error Diff'] = 0.00
for i in range(4,39):
    df['Squared Error Diff'][i] = "{:,.2f}".format((df['Error'][i] - df['Error'][i-1] )**2)
df.head()                                
```

```python
df.tail()
```

```python
MSE = np.mean(df['Squared Error'][3:-1])
MSE
```

```python
RMSE = np.sqrt(np.mean(df['Squared Error'][3:-1]))
RMSE
```

```python
MAD = np.mean(df['Absolute Error'][3:-1])
MAD
```

```python
MAPE = np.mean(df['Precent Error1'][3:-1])
MAPE
```

```python
ThielsU = np.sqrt( np.sum( df['Precent Error2'][3:-1] ) / np.sum( df['Precent Error3'][3:-1] ) )
ThielsU 
```
